# Render graph with graphviz (not pygraphviz) bindings

This sandbox example builds the graph directly as a `graphviz.Digraph()` object. Since those objects are only for rendering, and do not support, for example, general traversal, we need instead to build the graph first as a `networkx.DiGraph()` object and then copy it as a `graphviz.Digraph()` object for rendering. We do that in a different notebook.

## Reload libraries on each run, since weâ€™re modifying them

## Load libraries

```python
import networkx as nx
import graphviz
from IPython.display import SVG
from IPython.core.display import display
```

## Create sample graph in networkx

```python
G = nx.DiGraph()
G.add_node(0, label="Arthur")
G.add_node(1, label="Bedevere")
G.add_node(2, label="Arthur") # make sure we can handle multiple nodes with the same label
G.add_edge(0, 1)
G.add_edge(0, 2)
```

## Look around

```python
G.nodes()
```

```python
G.node[0]
```

```python
G.node[0]["label"]
```

```python
for id, attrs in G.nodes.items():
    print(attrs["label"])
```

```python
G.edges()
```

```python
for x in G.edges():
    print(x)
```

## Create graphviz graph

```python
G1 = graphviz.Digraph(format="svg")
```

```python
for id, attrs in G.nodes.items():
    G1.node(str(id), label=attrs["label"]) # node ids in graphviz must be str, not int
```

```python
G1 # just nodes so far
```

```python
for u, v in G.edges():
    G1.edge(str(u), str(v))
```

```python
G1 # full graph, after adding edges
```
